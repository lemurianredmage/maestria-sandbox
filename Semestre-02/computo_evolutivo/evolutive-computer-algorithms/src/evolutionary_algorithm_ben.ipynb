# Run evolutionary algorithm

```python
import numpy as np
import matplotlib.pyplot as plt

# Paso 1: Definir la función objetivo (fitness)
def rastrigin(x):
    A = 10
    return A * len(x) + sum([(xi**2 - A * np.cos(2 * np.pi * xi)) for xi in x])

# Paso 2: Definir el espacio de búsqueda y otras configuraciones
bounds = (-5.12, 5.12)
dimension = 10

# Paso 3: Crear y correr el algoritmo evolutivo
ea = CustomEvolutionaryAlgorithm(
    fitness_function=rastrigin,
    bounds=bounds,
    dimension=dimension,
    population_size=50,
    crossover_prob=0.9,
    mutation_prob=0.2,
    tournament_size=3,
    elitism=True,
    max_evals=500,
    max_age=5
)

best_individual, best_fitness, history = ea.run()

print(len(best_individual))
print(len(history))
# Paso 4: Mostrar resultados
print("Mejor solución encontrada:")
print("x =", best_individual)
print("Fitness =", best_fitness)

# Paso 5: Graficar convergencia (fitness vs evaluaciones)
evals = [item[0] for item in history]
fitness_values = [item[2] for item in history]

plt.plot(evals, fitness_values, label="Mejor fitness")
plt.xlabel("Evaluaciones de la función objetivo")
plt.ylabel("Fitness")
plt.title("Convergencia del algoritmo evolutivo")
plt.grid(True)
plt.legend()
plt.show()
```
